```python
!pip install yfinance hmmlearn matplotlib numpy pandas --quiet
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from hmmlearn.hmm import GaussianHMM
from datetime import datetime
from IPython.display import HTML
import matplotlib.animation as animation
symbol = "^GSPC"   # S&P 500 index
data = yf.download(symbol, start="2016-01-01", end="2025-01-01")
data.dropna(inplace=True)
data.head()
data['Returns'] = data['Close'].pct_change()
data['Volatility'] = data['Returns'].rolling(window=7).std()
data.dropna(inplace=True)

X = data[['Returns', 'Volatility']].values
n_states = 3  # Bull, Sideways, Bear
model = GaussianHMM(n_components=n_states, covariance_type='diag', n_iter=200, random_state=41)
model.fit(X)

data['State'] = model.predict(X)
probs = model.predict_proba(X)  # Probability of each state at each time
plt.figure(figsize=(14,8))

# Price & States
plt.subplot(3,1,1)
plt.title(f"Market Price & Hidden States – {symbol}")
for i, color in enumerate(['red', 'blue', 'green']):
    mask = data['State'] == i
    plt.plot(data.index[mask], data['Close'][mask], '.', color=color, label=f"State {i}")
plt.legend()

# Returns
plt.subplot(3,1,2)
plt.plot(data.index, data['Returns'], color='black', linewidth=0.7)
plt.title("Daily Returns")

# Volatility
plt.subplot(3,1,3)
plt.plot(data.index, data['Volatility'], color='orange', linewidth=0.8)
plt.title("7-Day Rolling Volatility")

plt.tight_layout()
plt.show()
print("State Transition Matrix:")
print(pd.DataFrame(model.transmat_,
                   columns=[f"To State {i}" for i in range(n_states)],
                   index=[f"From State {i}" for i in range(n_states)]))
# Prepare figure
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12,8), sharex=True,
                               gridspec_kw={'height_ratios':[2,1]})
fig.suptitle(f"Hidden Markov Model – Market Regimes ({symbol})", fontsize=14)

colors = ['red', 'blue', 'green']
labels = ['Bear', 'Sideways', 'Bull']

line, = ax1.plot([], [], lw=2, color='black')
ax1.set_ylabel("Price Level")
ax1.set_xlim(data.index.min(), data.index.max())
ax1.set_ylim(data['Close'].min().item()*0.98, data['Close'].max().item()*1.02)
text_label = ax1.text(0.02, 0.9, '', transform=ax1.transAxes, fontsize=12,
                      bbox=dict(facecolor='white', alpha=0.7))

# For probability lines
prob_lines = [ax2.plot([], [], color=c, label=l)[0] for c,l in zip(colors, labels)]
ax2.set_xlim(data.index.min(), data.index.max())
ax2.set_ylim(0,1)
ax2.set_ylabel("State Probability")
ax2.legend(loc="upper right")

def init():
    line.set_data([], [])
    for pl in prob_lines:
        pl.set_data([], [])
    text_label.set_text('')
    return [line, text_label] + prob_lines

def update(frame):
    if frame > 0:
        x = data.index[:frame]
        y = data['Close'].iloc[:frame]
        line.set_data(x, y)

        state_now = data['State'].iloc[frame-1]
        text_label.set_text(f"State: {labels[state_now]}")
        text_label.set_color(colors[state_now])

        for i, pl in enumerate(prob_lines):
            pl.set_data(data.index[:frame], probs[:frame, i])

        # Clear previous axvspan artists
        for collection in ax1.collections:
            collection.remove()

        ax1.axvspan(x[-1], x[-1], color=colors[state_now], alpha=0.15)

    return [line, text_label] + prob_lines

ani = animation.FuncAnimation(fig, update, frames=len(data), init_func=init,
                              blit=False, interval=40)

plt.close()
HTML(ani.to_jshtml())
```
